# Visualizing movies per year

```python
plt.figure(figsize=(12,4))
plt.title('Number of shows increases per year',fontdict={'fontsize':25},fontweight="bold")
plt.xlabel('Year')
plt.ylabel('Count')
sns.lineplot(x=hotdata.year.value_counts().index, y=hotdata.year.value_counts(), color='r')
     
```

Number of Movies TV shows per Year

```python
plt.figure(figsize=(80,40))
plt.title("Number of Shows over the year",fontdict={'fontsize':200})
plt.ylabel("Number of Movies/TV Shows",fontdict={'fontsize':100})
plt.xlabel("Year",fontdict={'fontsize':100})
sns.barplot(x=hotdata.year.value_counts().index, y=hotdata.year.value_counts(), color='r',palette='viridis')
```

Top 10 Genres

```python
plt.figure(figsize=(16,8))
plt.title("Top Ten Genres  ",fontdict={'fontsize':20})
plt.xlabel("Genres",fontdict={'fontsize':20})
plt.ylabel("Count",fontdict={'fontsize':20})
plt.xticks(rotation=0)
data=hotdata.genre.value_counts()[:10]
sns.barplot(x=data.index,y=data)
```

Verfifying distibution of runnig time by boxplot

```python
sns.boxplot(x=hotdata["running_time"])
```

```python
#calculating the quartiles 
Quartile_1 = hotdata.running_time.quantile(.25)
Quartile_3 = hotdata.running_time.quantile(.75)
inner_Quartile=Quartile_3-Quartile_1
```

```python
print("The Quartile 1 is :",Quartile_1,",The Quartile 3 is :",Quartile_3,",The inner quartile is :",inner_Quartile)
```

Age Ratings of Hotstar Shows

```python
data= hotdata.age_rating.value_counts()
plt.figure(figsize=(16,8))
sns.set_style('darkgrid')
plt.title('Age Rating of Hotstar Shows',fontdict={'fontsize':25},fontweight="bold")
plt.xlabel('Age Rating',fontdict={'fontsize':15})
plt.xticks(rotation=0)
plt.ylabel('Count',fontdict={'fontsize':15})
sns.barplot(x=data.index,y=data)
     
```

```python
data = hotdata.type.value_counts()
plt.pie(data,labels=data.index,startangle=90,autopct='%1.1f%%')
```

Most Common Words in the dataset

```python
from wordcloud import WordCloud
image1 = " ".join(i for i in hotdata['genre'])
word_cloud1 = WordCloud(collocations = True, background_color = 'white',
                        width = 1920, height = 1080).generate(image1)
word_cloud1.to_file('ima1.png')
plt.figure(figsize=(16,16))
plt.imshow(word_cloud1, interpolation='bilinear')
plt.axis("off")
plt.show()
```
